*Etapa 3*: : Inicializar a Matriz de Contratos

Depois de obter os valores de `m` e `n`, precisamos criar uma matriz tridimensional para armazenar os valores dos contratos. A matriz será composta de:

1. **Primeira dimensão**: Fornecedores.
2. **Segunda dimensão**: Mès de nício do contrato.
3. **Terceira dimensão**: Mês de término do contrato.



**Abaixo é apresentado como as informações foramserão agregadas na matriz 3D**

<img src="https://github.com/pedrorela/MBA_Introducao_Python/blob/main/3dbox.jpg?raw=true" width="400">

```r
# Função para exibir tipo, classe e tamanho
info_obj <- function(obj) {
    classe <- class(obj)
    tipo <- typeof(obj)
    tam <- object.size(obj)

    if (is.matrix(obj) | is.array(obj) | is.list(obj)) {
        print(obj)
        cat(classe, "dim:", dim(obj), tipo, tam, "bytes\n")
    } else {
        cat(obj, classe, tipo, tam, "bytes\n")
    }
}
```

```r
# Num.Row <- Mes_Max_Fim 
# Num.Col <- Mes_Max_Ini
# Profundidade  <- df_Num_fornecedor_2

#contratos[inicio][fim][fornecedor]


contratos <- array(dim = c(Mes_Max_Ini, Mes_Max_Fim, df_Num_fornecedor_2))
contratos[] <- Inf

info_obj(contratos)

```

```r
# Preenchendo o array com os valores do data frame

for (row in 1:nrow(df)) {
  
  mes_ini <- df$Mes_Inicio[row]
  mes_fim <- df$Mes_Fim[row]
  fornecedor <- df$Fornecedor[row]
  valor <- df$Valor[row]

  contratos[mes_ini,mes_fim , fornecedor] <- valor

}

print(contratos)
```
